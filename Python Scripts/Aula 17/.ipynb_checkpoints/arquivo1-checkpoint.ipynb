```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def face_detector(imagem):
    face_imagem = imagem.copy()

    face_retangulo = face_cascade.detectMultiScale(face_imagem, scaleFactor = 1.2, minNeighbors = 3)

    for x, y, w, h in face_retangulo:
        cv2.rectangle(face_imagem, (x, y), (x + w, y + h), (255, 0, 0), 10)

    return face_imagem
```

```python
face_cascade = cv2.CascadeClassifier("haarcascade/haarcascade_frontalface_default.xml")
```

```python
imagem1 = cv2.imread("imagem.png")
imagem1 = cv2.cvtColor(imagem1, cv2.COLOR_BGR2GRAY)

imagem2 = cv2.imread("pessoas.png")
imagem2 = cv2.cvtColor(imagem2, cv2.COLOR_BGR2GRAY)
```

```python
resultado = face_detector(imagem1)
plt.imshow(resultado, cmap = "gray")
plt.show()
```

```python
resultado = face_detector(imagem2)
plt.imshow(resultado, cmap = "gray")
plt.show()
```
