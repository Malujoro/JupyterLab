```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def face_detector(imagem):
    face_imagem = imagem.copy()

    face_retangulo = face_cascade.detectMultiScale(face_imagem, scaleFactor = 1.2, minNeighbors = 3)

    for x, y, w, h in face_retangulo:
        cv2.rectangle(face_imagem, (x, y), (x + w, y + h), (255, 0, 0), 10)

    cv2.rectangle(face_imagem, (0, 0), (50, 40), (0, 0, 0), -1)
    cv2.putText(face_imagem, f"{len(face_retangulo)}", (0, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
    return face_imagem
```

```python
face_cascade = cv2.CascadeClassifier("haarcascade/haarcascade_frontalface_default.xml")
```

```python
video = cv2.VideoCapture(0)

while True:

    ret, frame = video.read()

    frame = face_detector(frame)
    cv2.imshow("tela", frame)
    
    if(cv2.waitKey(20) & 0xFF == 27):
        break

video.release()
cv2.destroyAllWindows()
```
