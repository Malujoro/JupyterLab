```python
import numpy as np
import cv2
```

```python
name = 'minha_tela'
# Desenho é Falso quando o clique não estiver acionado
# Desenho é Verdadeiro quando o clique estiver pressionado
desenho = False
ix = -1
iy = -1


def retangulo(event, x, y, flags, params):
    global ix, iy, desenho

    # Quando o botão esquerdo é clicado
    if(event == cv2.EVENT_LBUTTONDOWN): 
        ix = x
        iy = y
        desenho = True

    # Quando o mouse está clicado e o mouse sendo movido
    elif(event == cv2.EVENT_MOUSEMOVE and desenho):
        cv2.rectangle(imagem, (ix, iy), (x, y), (255, 255, 255), -1)
        
    elif(event == cv2.EVENT_LBUTTONUP):
        desenho = False
        cv2.rectangle(imagem, (ix, iy), (x, y), (0, 255, 255), -1)

cv2.namedWindow(winname = name)
cv2.setMouseCallback(name, retangulo)

imagem = np.zeros((400, 400, 3))

while True:

    cv2.imshow(name, imagem)

    if(cv2.waitKey(20) & 0xFF == 27):
        break

cv2.destroyAllWindows()
```
