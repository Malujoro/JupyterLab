```python
import cv2

def desenhaRetangulo(event, x, y, flags, params):

    global pt1, pt2, topo, base

    if(event == cv2.EVENT_LBUTTONDOWN):

        if(topo and base):
            pt1 = (0, 0)
            pt2 = (0, 0)
            topo = False
            base = False

        elif(not topo):
            pt1 = (x, y)
            topo = True
            
        elif(not base):
            pt2 = (x, y)
            base = True

pt1 = (0, 0)
pt2 = (0, 0)
topo = False
base = False
name = "frame"

video = cv2.VideoCapture(0)

cv2.namedWindow(name)
cv2.setMouseCallback(name, desenhaRetangulo)

while True:

    ret, frame = video.read()

    if(topo):
        cv2.circle(frame, center = pt1, radius = 3, color = (0, 255, 0), thickness = -1)

        if(base):
            cv2.rectangle(frame, pt1, pt2, color = (0, 0, 0), thickness = 6)
    
    cv2.imshow(name, frame)
    
    if(cv2.waitKey(1) & 0xFF == ord('s')):
        break

video.release()
cv2.destroyAllWindows()
```
