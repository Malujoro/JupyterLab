```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
xadrez = cv2.imread("xadrez.png")
xadrez = cv2.cvtColor(xadrez, cv2.COLOR_BGR2RGB)
cinza_xadrez = cv2.cvtColor(xadrez, cv2.COLOR_RGB2GRAY)
plt.imshow(xadrez)
# plt.imshow(cinza_xadrez, cmap = "gray")
plt.show()
```

```python
bordas = cv2.goodFeaturesToTrack(cinza_xadrez, 48, 0.01, 10)
```

```python
bordas = np.intp(bordas)
```

```python
copia_xadrez = xadrez.copy()
for i in bordas:
    x, y = i.ravel()

    cv2.circle(copia_xadrez, (x, y), 3, (0, 255, 0), -1)

plt.imshow(copia_xadrez)
plt.show()
```

```python
xadrez_real = cv2.imread("xadrez_real.png")
xadrez_real = cv2.cvtColor(xadrez_real, cv2.COLOR_BGR2RGB)
cinza_xadrez_real = cv2.cvtColor(xadrez_real, cv2.COLOR_RGB2GRAY)
plt.imshow(xadrez_real)
plt.imshow(cinza_xadrez_real, cmap = "gray")
plt.show()
```

```python
bordas = cv2.goodFeaturesToTrack(cinza_xadrez_real, 100, 0.01, 10)
```

```python
bordas = np.intp(bordas)
```

```python
copia_xadrez_real = xadrez_real.copy()
for i in bordas:
    x, y = i.ravel()

    cv2.circle(copia_xadrez_real, (x, y), 7, (0, 255, 0), -1)

plt.imshow(copia_xadrez_real)
plt.show()
```
