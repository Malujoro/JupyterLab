```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
imagem = cv2.imread("lena.png")
```

```python
type(imagem)
```

```python
plt.imshow(imagem)
plt.show()
```

```python
nova_imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(nova_imagem)
plt.show()
```

```python
nova_imagem.shape
```

```python
redimensionada_imagem = cv2.resize(nova_imagem, (300, 250))
```

```python
plt.imshow(redimensionada_imagem)
plt.show()
```

```python
redimensionada_imagem.shape
```

```python
largura = 0.5
altura = 0.3
```

```python
outra_imagem = cv2.resize(nova_imagem, (0,0), nova_imagem, largura, altura)
```

```python
plt.imshow(outra_imagem)
plt.show()
```

```python
# Girar a imagem
# 0 vira de ponta cabea
# -1 "Espelha" verticalmente
imagem_girada = cv2.flip(nova_imagem, -1)
```

```python
plt.imshow(imagem_girada)
plt.show()
```

```python
# Criar uma cpia da figura / salvar ela
cv2.imwrite("lena_modificado.png", imagem)
```

```python
cv2.imwrite("lena_girada.png", imagem_girada)
```

```python
# Aumenta / diminui o tamanho da imagem exibida
img = plt.figure(figsize=(10,8))
ax = img.add_subplot(111)
ax.imshow(imagem_girada)
plt.show()
```
