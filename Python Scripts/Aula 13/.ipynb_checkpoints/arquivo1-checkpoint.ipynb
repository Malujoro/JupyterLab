```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
xadrez = cv2.imread("xadrez.png")
xadrez = cv2.cvtColor(xadrez, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(xadrez)
plt.show()
```

```python
cinza_xadrez = cv2.cvtColor(xadrez, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(cinza_xadrez, cmap = "gray")
plt.show()
```

```python
xadrez_real = cv2.imread("xadrez_real.png")
xadrez_real = cv2.cvtColor(xadrez_real, cv2.COLOR_BGR2RGB)
plt.imshow(xadrez_real)
plt.show()
```

```python
cinza_xadrez_real = cv2.cvtColor(xadrez_real, cv2.COLOR_RGB2GRAY)
plt.imshow(cinza_xadrez_real, cmap = "gray")
plt.show()
```

```python
cinza_xadrez
```

```python
cinza_xadrez_real
```

```python
cinza = np.float32(cinza_xadrez)
```

```python
cinza
```

```python
destino = cv2.cornerHarris(src = cinza_xadrez, blockSize = 2, ksize = 3, k = 0.04)
```

```python
destino = cv2.dilate(destino, None)
```

```python
xadrez[destino > 0.01 * destino.max()] = [0, 255, 0]
```

```python
destino.max()
```

```python
destino
```

```python
plt.imshow(xadrez)
plt.show()
```

```python
cinza_real = np.float32(cinza_xadrez_real)
```

```python
destino_real = cv2.cornerHarris(src = cinza_xadrez_real, blockSize = 2, ksize = 3, k = 0.04)
```

```python
destino_real = cv2.dilate(destino_real, None)
```

```python
xadrez_real[destino_real > 0.01 * destino_real.max()] = [255, 0, 0]
```

```python
plt.imshow(xadrez_real)
plt.show()
```
