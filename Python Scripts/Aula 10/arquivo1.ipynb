```python
import cv2

video = cv2.VideoCapture("video.mp4")

if(not video.isOpened()):
    print("Erro ao abrir o vídeo")


while video.isOpened():

    ret, frame = video.read()

    if(ret):
        cv2.imshow("frame", frame)

        if(cv2.waitKey(30) & 0xFF == ord('s')):
            break
    else:
        break

video.release()
cv2.destroyAllWindows()
```
