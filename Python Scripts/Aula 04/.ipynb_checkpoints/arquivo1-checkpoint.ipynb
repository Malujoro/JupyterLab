```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from PIL import Image
```

```python
# Abre a imagem
imagem = Image.open('lena.png')
```

```python
type(imagem)
```

```python
# Cria um vetor baseado na imagem
imagemVetor = np.asarray(imagem)
```

```python
type(imagemVetor)
```

```python
imagemVetor.shape
```

```python
# Exibir a imagem
plt.imshow(imagemVetor)
plt.show()
```

```python
# Copiar o vetor
copiaImagem = imagemVetor.copy()
```

```python
type(copiaImagem)
```

```python
plt.imshow(copiaImagem)
plt.show()
```

```python
copiaImagem.shape
```

```python
# Exibir a imagem, pode ser passado o "mapa de cor" (cmap)
plt.imshow(copiaImagem[:, :, 0], cmap = 'gray')
plt.show()
```

```python
copiaImagem[:, :, 2]
```

```python
copiaImagem[:, :, 2] = 0
```

```python
plt.imshow(copiaImagem)
plt.show()
```
